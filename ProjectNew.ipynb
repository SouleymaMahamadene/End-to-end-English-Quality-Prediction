## Normalization of data

```python
df_modified = df.copy()
df_modified['averageD1'] = 0
```

```python
for i in range(1,9):
    df_max = df_modified.domain1_score.loc[(df_modified['essay_set'] == i)].max()
    df_modified.averageD1.loc[(df_modified['essay_set'] == i)] = df_modified.domain1_score.loc[(df_modified['essay_set'] == i)]/df_max
    df_modified['categoryD1'] = pd.cut(df_modified['averageD1'], bins=[-float('inf'), 0.42, 0.67, float('inf')], labels=['bad', 'good', 'excellent'])
    
    if i == 2:
        df_max = df_modified.domain2_score.loc[(df_modified['essay_set'] == i)].max()
        df_modified['averageD2'] = df_modified.domain2_score.loc[(df_modified['essay_set'] == i)]/df_max
        df_modified['categoryD2'] = pd.cut(df_modified['averageD2'], bins=[-float('inf'), 0.51, 0.75, float('inf')], labels=['bad', 'good', 'excellent'])
    
df_modified
```

### Redability metrics 
Redability metrics give an estimate of the text. They correspond to the U.S grade level of education which varies from 0-18  

#### Flesch-Kincaid Grade Level

```python
df_ws['FK'] = 0
```

```python
for i in range(len(df_ws)):
    text = df_ws.essay[i]
    if (df_ws.Num_of_words[i] > 101):
        r = Readability(text)
        fk = r.flesch_kincaid()
        df_ws.FK[i] = fk.score
df_ws
```

#### Flesch Reading Ease

```python
for i in range(len(df_ws)):
    text = df_ws.essay[i]
    if (df_ws.Num_of_words[i] > 101):
        r = Readability(text)
        f = r.flesch()
        df_ws.FRE[i] = f.score
df_ws
```

#### Simple Measure of Gobbledygook (SMOG)

```python
from re import split
```

```python
df_ws['Num_of_sentences'] = 0
df_ws['SMOG'] = 0
```

```python
for i in range(len(df_ws)):
    text = df_ws.essay[i]
    df_ws.Num_of_sentences[i] = sum(any(map(str.isalpha, s)) for s in split(r'[.!?]', text))
df_ws
```

```python
for i in range(len(df_ws)):
    text = df_ws.essay[i]
    num = df_ws.Num_of_sentences[i]
    if num >= 31:
        r = Readability(text)
        s = r.smog()
        df_ws.SMOG[i]= s.score
df_ws
```
